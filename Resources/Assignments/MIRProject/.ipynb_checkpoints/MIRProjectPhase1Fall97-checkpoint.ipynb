<br>
عملیات مورد انتظار:
<ol>
    <li>
یکسان‌سازی متن: یکی از عملیات مهم در پردازش متون به خصوص در زبان فارسی این عملیات
است که شامل یکسان‌سازی استفاده از فاصله و نیم‌فاصله و نحوه‌ی شکستن یا ادغام کلمات و ... است.  به طور مثال، یک مورد از این یکسان‌سازی‌ها نحوه‌ی قرار گیری حرف جمع «ها» در انتهای کلمات جمع است که می‌تواند بدون فاصله چسبیده به کلمه، با یک فاصله‌ی کامل و یا با نیم‌فاصله
پس از کلمه بیاید (کتابها، کتاب ها، کتاب‌ها)
    </li>
    <li> 
جدا کردن کلمات یک جمله: واحد متن مورد استفاده‌ی ما در ساخت نمایه و همین طور جست‌وجو در یک سیستم اطلاعاتی کلمات هستند. بنابر این جملات ورودی را باید بتوانیم به کلمات آن بشکنیم  و عملیات مورد نیاز را بر روی کلمات انجام دهیم.
    </li>
    <li>
حذف علائم نگارشی: علائم نگارشی مانند نقطه، ویرگول، و ... باید از درون اسناد حذف شوند تا درون نمایه و جست‌وجو‌ها تاثیر نگذارند.
    </li>
    <li>
حذف کلمات پرتکرار: یکی دیگر از عملیات مهم بر روی متون در سیستم‌های بازیابی اطلاعات،
حذف کلمات پرتکرار است چرا که این کلمات اطلاعات زیادی را در خصوص مفهوم مورد جست‌وجو به  ما نمی‌دهند و همین طور به دلیل تعداد زیادشان حجم زیادی را در ساخت نمایه اشغال می‌کنند. به همین جهت کلمات پرتکرار از محتوای اسناد و پرسمان‌ها حذف می‌شوند. شما می‌بایست ابتدا کلمات پرتکرار مناسب برای این مجموعه داده را پیدا کرده و سپس برای آماده‌سازی اولیه از این کلمات برای حذف کلمات پر کاربرد استفاده کنید.

</li>
<li>
بازگرداندن کلمات به ریشه: عملیات دیگری که روی کلمات متن صورت میگیرد عمل بازگردانی به
ریشه
(stemming)
است تا کلماتی که از یک ریشه هستند همگی یک کلمه به حساب بیاید.
    </li>
</ol>

در این بخش دو شکل پرسمان وجود دارد:
<ol>
<li>
پرسمان کلی: که جست‌وجوی یک پرسمان بر روی هر دو بخش عنوان و توضیح صورت می‌گیرد و سپس نتیجه بر اساس امتیاز وزن‌دار جست و جو بر روی این دو بخش به ترتیب برگردانده می‌شود. وزن امتیاز جست‌وجو در عنوان نسبت به وزن امتیاز جست‌وجو در توضیح باید به عنوان پارامتر ورودی قابل تنظیم باشد اما در حالت پیش‌فرض آن را ۲ در نظر می‌گیریم. 
</li>
<li>
پرسمان بر اساس بخش: که به عنوان ورودی دو پرسمان یکی برای عنوان و یکی برای توضیح می‌آید و جست و جو برای هر بخش به طور مجزا صورت گرفته و در نهایت امتیاز به دست آمده برای جست و جوی هر بخش به شکل مساوی با هم جمع شده و نتیجه به ترتیب برگردانده می‌شود.
</li>
</ol>
برای هر دو نوع پرسمان فوق، پس از مشخص شدن روش امتیازدهی به عنوان ورودی
(ltn-lnn
و
ltc-lnc)
شما باید لیستی مرتب از شناسه اسناد بر اساس امتیاز کسب شده برگردانید. امتیازات بر اساس توضیحات بالا باید محاسبه شوند. همین طور در هر دو نوع پرسمان، به عنوان ورودی لیستی از دسته‌های کالا‌ها داده می‌شود که در صورت خالی نبودن لیست،  جست‌وجو باید تنها در کالاهای درون دسته‌های مشخص شده در ورودی صورت بگیرد و در صورت خالی بودن لیست جست و جو بدون در نظر گرفتن فیلتر دسته‌بندی بر روی تمامی کالا‌ها صورت پذیرد.

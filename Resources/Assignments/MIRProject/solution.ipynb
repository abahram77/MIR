```python
punctuations_fa = ['.', '؟', '،', '!', '؛', '-', '_', '+', '=', '﷼']
punctuations_en = [',', '\"', "\'", ':', '{', '}', ')', '(', '\\']
punctuations = []
punctuations.extend(punctuations_en)
punctuations.extend(punctuations_fa)



class NLP:
    
    def __init__(self, path):
        self.path = path
        self.tokens = []
        
    def stopwords(self):
        stopword = []
        # TODO:
        return stopword

    def read_file(self):
        normalizer = Normalizer()
        stemmer = Stemmer()
        with open(self.path, 'r') as f:
            for line in f.readlines():
                for punc in punctuations:
                    line = line.replace(punc, '')
                normal_line = normalizer.normalize(line)
                tokens = word_tokenize(normal_line)
                for token in tokens:
                    stemmed_token = word_tokenize(token)
                    self.tokens.append(stemmed_token)
                
nlp = NLP('./data/adds.json')
text = nlp.read_file()
```
